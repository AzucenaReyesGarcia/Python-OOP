# Ejercicio 1

<img src=".Diagramas_clases/01_create_class.png" alt="Ex 1" with="500"

```python
class Employee:
    new_id = 1
    
    def __init__(self, name):
        self.id = Employee.new_id
        self.name = name
        Employee.new_id += 1
        
    def say_id(self): 
        print(f'Mi nombre es  {self.name} y mi id es: {self.id}')
```

```python
Asael_Luces = Employee("Asael")
Citlali_Traidora = Employee("Citlali")

Asael_Luces.say_id() 
Citlali_Traidora.say_id() 
```

# Ejercicio 2


<img src=".Diagramas_clases/02_.png" alt="Ex 1" with="500"

```python
class Animal:
    def __init__(self, name):
        self.name = name  
        
    def say_name(self): 
        print(f'Mi nombre es: {self.name}')
    
    def eat(self): 
        print(f'Jummy Yummy!, estoy comiendo')
        
class Cat(Animal):
    def meow(self):
        print('Soy un Gato y hago meow!')
    
class Dog(Animal):
    def bark(self):
        print('Soy un Perro y hago gua gua!')
```

```python
Solo_Vino = Dog('Solo Vino')
Solo_Vino.bark()
Solo_Vino.eat()
Solo_Vino.say_name()

Felix = Cat('Felix')
Felix.meow()
Felix.eat()
Felix.say_name()
```
